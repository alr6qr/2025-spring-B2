When comparing a **histogram** and a **boxplot** to visualize the distribution of drug prices, both offer pros and cons depending on the context and the insights you want to communicate.



---




### **Histogram**
**Pros**:
1. **Detailed Frequency Distribution**: Histograms provide a clear view of how many drugs fall into specific price ranges. This makes it easier to see where most drug prices cluster, especially useful for identifying modes or peak price ranges.
2. **Visualizing Skewness**: If the prices are skewed (for example, a few very expensive drugs), the histogram can make this pattern obvious by showing how the data is distributed across different price bins.

**Cons**:
1. **Loss of Key Summary Information**: While histograms are great for seeing the distribution, they don’t directly summarize important values like the median, quartiles, or outliers, which might be important when trying to understand the "more expensive" drugs.
2. **Sensitive to Bin Size**: The appearance of a histogram can vary greatly depending on the number of bins used. Too many bins might make it hard to see patterns, while too few could obscure meaningful details.


### **Boxplot**

**Pros**:
1. **Quick Summary of Data**: A boxplot succinctly summarizes the key statistics—**median**, **interquartile range (IQR)**, **minimum**, **maximum**, and **outliers**—which is particularly useful when identifying how much the more expensive drugs cost in comparison to others.

2. **Outlier Detection**: Boxplots are excellent for quickly spotting outliers (extremely expensive drugs), which can be valuable when the administrator is interested in distinguishing outlier drugs from the rest of the distribution.
3. **Focus on Central Tendency and Spread**: Boxplots highlight the median and quartiles, giving a clear picture of where most data points lie and how they are spread.

**Cons**:
1. **Less Detail on Distribution**: While it shows the spread and outliers, a boxplot doesn’t convey detailed frequency data. You can’t see how many drugs fall into specific price ranges, which might be important for some analysis.
2. **No Visual on Shape**: A boxplot doesn't show the exact shape of the distribution (whether it's skewed or symmetric), so if understanding the overall shape of the price distribution is important, the boxplot won’t give that information.



---




**Recommendation:**

 **My choice would be to use the boxplot.** Given the question of how much the **more expensive drugs** cost, in the context of other prices, a **boxplot** is the better choice. The boxplot provides a clear visualization of the **median price** and **outliers**, making it to easier to identify where the more expensive drugs fall in relation to the rest of the data.


---



**Boxplot Visual Enhancements**

   - **Color**: You could color the boxplot with a gradient to show how prices increase, making it visually more appealing and highlighting trends.

   - **Annotations**: Add labels for the **median**, **Q1**, **Q3**, and any significant outliers, so viewers can immediately grasp the important values.

   - **Outlier Symbols**: Use distinct symbols like circles or stars for outliers to make them stand out more clearly.


```python
import matplotlib.pyplot as plt

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size=num_data) * 4)
df = pd.DataFrame(data.T, columns=["data"])
print("The 75th percentile is:", df.quantile(q=0.75))

# Create the boxplot
plt.figure(figsize=(8, 6))  # Adjust figure size for better visualization
boxplot = df.boxplot(column='data', patch_artist=True, showfliers=True,
                     flierprops=dict(marker='o', markerfacecolor='red', markersize=8, markeredgecolor='black'))

# Color the boxplot with a gradient
colors = ['lightblue', 'skyblue', 'steelblue']  # Example gradient colors
#boxes = [patch for patch in boxplot.patches if type(patch) == matplotlib.patches.PathPatch]
#for patch, color in zip(boxes, colors):
    #patch.set_facecolor(color)

# Annotate key statistics
median = df['data'].median()
q1 = df['data'].quantile(0.25)
q3 = df['data'].quantile(0.75)

plt.text(1.05, median, f'Median: {median:.2f}', va='center')
plt.text(1.05, q1, f'Q1: {q1:.2f}', va='center')
plt.text(1.05, q3, f'Q3: {q3:.2f}', va='center')

# Customize outlier symbols
# Get the fliers from the boxplot object
flier = boxplot.lines[0]
# Get the x and y data from the fliers
outlier_x = flier.get_xdata()
outlier_y = flier.get_ydata()

for x,y in zip(outlier_x, outlier_y):
    plt.plot(x, y, marker='*', markersize=10, markerfacecolor='purple', markeredgecolor='black')


plt.title('Boxplot of Drug Prices')
plt.ylabel('Price')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the CSV file
file_path = 'amazon.csv'

try:
    df = pd.read_csv(file_path)
    print(df.head())
    print(df.info())

    # Data Cleaning: Convert necessary columns to numeric, checking if they exist first
    if 'rating' in df.columns:
      df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
    if 'rating_count' in df.columns:
        df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
    if 'total_rating_count' in df.columns:
      df['total_rating_count'] = pd.to_numeric(df['total_rating_count'], errors='coerce')

    # 1. Draw Histograms for Each Numeric Feature
    def draw_histograms(df):
        # Loop through each numeric column
        for column in df.select_dtypes(include='number').columns:
            plt.figure(figsize=(8, 6))
            sns.histplot(df[column], kde=True, bins=40)
            plt.title(f'Histogram and KDE for {column}')
            plt.show()

    # 2. Draw Histogram Variants (Swarm Plot, KDE Plot, Violin Plot) for Each Numeric Feature
    def draw_histogram_variants(df):
        for column in df.select_dtypes(include='number').columns:
            # Swarm plot
            plt.figure(figsize=(8, 6))
            sns.swarmplot(x=df[column])
            plt.title(f'Swarm Plot for {column}')
            plt.show()

            # KDE plot
            plt.figure(figsize=(8, 6))
            sns.kdeplot(df[column], fill=True)
            plt.title(f'KDE Plot for {column}')
            plt.show()

            # Violin plot
            plt.figure(figsize=(8, 6))
            sns.violinplot(x=df[column])
            plt.title(f'Violin Plot for {column}')
            plt.show()

    # 3. Draw Grouped Histograms (Example: Price for Different Categories)
    def draw_grouped_histograms(df, numeric_column, categorical_column):
        plt.figure(figsize=(10, 6))
        sns.histplot(data=df, x=numeric_column, hue=categorical_column, kde=True, bins=20, multiple="stack")
        plt.title(f'Grouped Histograms for {numeric_column} by {categorical_column}')
        plt.show()

    # 4. Draw Bar Plot for Categorical Variables (Category Distribution)
    def draw_bar_plot(df, categorical_column):
        plt.figure(figsize=(8, 6))
        sns.countplot(data=df, x=categorical_column)
        plt.title(f'Bar Plot for {categorical_column}')
        plt.show()

    # 5. Find Means, Medians, and Modes for Numeric Columns
    def calculate_statistics(df):
        numeric_cols = df.select_dtypes(include='number').columns
        means = df[numeric_cols].mean()
        medians = df[numeric_cols].median()
        modes = df[numeric_cols].mode().iloc[0]  # mode returns a series, so we take the first one
        print(f"Means:\n{means}\n")
        print(f"Medians:\n{medians}\n")
        print(f"Modes:\n{modes}\n")

    # Call the functions on your dataframe df
    draw_histograms(df)
    draw_histogram_variants(df)
    # Assuming columns are 'discounted_price' and 'category' and ensuring they exist
    if 'discounted_price' in df.columns and 'category' in df.columns:
      draw_grouped_histograms(df, 'discounted_price', 'category')
    # Assuming categorical column 'category' exists
    if 'category' in df.columns:
        draw_bar_plot(df, 'category')
    calculate_statistics(df)



except FileNotFoundError:
    print(f"Error: File '{file_path}' not found. Please check the file path.")
except pd.errors.EmptyDataError:
    print(f"Error: File '{file_path}' is empty.")
except pd.errors.ParserError:
    print(f"Error: Unable to parse file '{file_path}'. Please check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Draw grouped histograms top 3 products.


import matplotlib.pyplot as plt
import seaborn as sns

def draw_grouped_histograms_top_products(df, numeric_column, product_column='product_name', top_n=3):
    """
    Draws grouped histograms for the top N products based on a numerical column.

    Args:
        df: The DataFrame containing the data.
        numeric_column: The name of the numerical column.
        product_column: The name of the product column.
        top_n: The number of top products to consider.
    """

    # 1. Find the top N products based on the frequency of the product_name
    top_products = df[product_column].value_counts().nlargest(top_n).index

    # 2. Filter the DataFrame to include only the top products
    top_products_df = df[df[product_column].isin(top_products)]

    # 3. Create grouped histograms
    plt.figure(figsize=(12, 6))
    sns.histplot(data=top_products_df, x=numeric_column, hue=product_column, multiple="dodge", kde=False, bins=20)
    plt.title(f'Grouped Histograms of {numeric_column} for Top {top_n} Products')
    plt.xlabel(numeric_column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
    plt.tight_layout() # Adjust subplot params for a tight layout.
    plt.show()

if 'discounted_price' in df.columns and 'product_name' in df.columns:
    draw_grouped_histograms_top_products(df, 'discounted_price')
else:
    print("Error: 'discounted_price' or 'product_name' columns not found in the DataFrame.")
```

```python
# Draw grouped histogram for bottom 3 products

import matplotlib.pyplot as plt
def draw_grouped_histograms_bottom_products(df, numeric_column, product_column='product_name', bottom_n=3):
    """
    Draws grouped histograms for the bottom N products based on a numerical column.

    Args:
        df: The DataFrame containing the data.
        numeric_column: The name of the numerical column.
        product_column: The name of the product column.
        bottom_n: The number of bottom products to consider.
    """

    # 1. Find the bottom N products based on the frequency of the product_name
    bottom_products = df[product_column].value_counts().nsmallest(bottom_n).index

    # 2. Filter the DataFrame to include only the bottom products
    bottom_products_df = df[df[product_column].isin(bottom_products)]

    # 3. Create grouped histograms
    plt.figure(figsize=(12, 6))
    sns.histplot(data=bottom_products_df, x=numeric_column, hue=product_column, multiple="dodge", kde=False, bins=20)
    plt.title(f'Grouped Histograms of {numeric_column} for Bottom {bottom_n} Products')
    plt.xlabel(numeric_column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
    plt.tight_layout() # Adjust subplot params for a tight layout.
    plt.show()

# Example usage (replace 'discounted_price' with the actual column name):
if 'discounted_price' in df.columns and 'product_name' in df.columns:
    draw_grouped_histograms_bottom_products(df, 'discounted_price')
else:
    print("Error: 'discounted_price' or 'product_name' columns not found in the DataFrame.")
```
